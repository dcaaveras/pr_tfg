## LDA implementation

```python
import pandas as pd
import gensim
from gensim.models import CoherenceModel
from gensim.corpora import Dictionary
import scipy.sparse as sparse
from gensim import matutils
import scipy
from tqdm import tqdm
import gensim.corpora as corpora
```

```python
news_tfidf = tfidf_noticias
corpus_news = matutils.Sparse2Corpus(scipy.sparse.csc_matrix(news_tfidf.transpose()))

mails_tfidf = tfidf_correos
corpus_mails = matutils.Sparse2Corpus(scipy.sparse.csc_matrix(mails_tfidf.transpose()))
```

##### Hyperparameter Tuning

```python
id2word = corpora.Dictionary.from_corpus(corpus_news, id2word=None)
```

```python
texts = [[id2word[word_id] for (word_id, tfidf_score) in doc] for doc in corpus_news]
```

```python
from tqdm import tqdm
# Definir una lista de posibles números de tópicos
min_topics = 2
max_topics = 30
step_size = 1
num_topics_list = range(min_topics, max_topics, step_size)

# crear una lista de modelos LDA y medidas de coherencia C_V para cada número de tópicos
lda_models = []
cv_scores = []
perplexity_scores = []
umass_scores = []
uci_scores = []

for num_topics in tqdm(num_topics_list):
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus_news,
                                            id2word=id2word,
                                            num_topics=num_topics,
                                            random_state=42,
                                            update_every=1,
                                            passes=40,
                                            iterations=200,chunksize = 10000,
                                            per_word_topics=True)
    lda_models.append(lda_model)
    coherence_model_cv = CoherenceModel(model=lda_model, texts=texts, dictionary=id2word, coherence='c_v')
    coherence_model_umass = CoherenceModel(model=lda_model, corpus=corpus_news, coherence='u_mass')
    coherence_model_uci = CoherenceModel(model=lda_model, texts=texts, dictionary=id2word, coherence='c_uci')

    cv_score = coherence_model_cv.get_coherence()
    umass_score = coherence_model_umass.get_coherence()
    uci_score = coherence_model_uci.get_coherence()
    
    cv_scores.append(cv_score)
    umass_scores.append(umass_score)
    perplexity_scores.append(lda_model.log_perplexity(corpus_news))
    uci_scores.append(uci_score)
    
    
    print("Num topics: ", num_topics,". Coherence: ", cv_score)
    


```

```python
# guardamos metricas
# cv_scores, umass_scores, perplexity_scores, uci_scores

import pandas as pd

# Crear un diccionario con las listas
data = {
    "n_topicos":num_topics_list,  
    'cv_scores': cv_scores,
    'umass_scores': umass_scores,
    'perplexity_scores': perplexity_scores,
    'uci_scores': uci_scores
}

# Crear el dataframe
df = pd.DataFrame(data)

# Guardar el dataframe como un archivo CSV
df.to_csv('metricas_modelo/v1/metricas_mainv1.csv', index=False)
```

```python
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
import math
from collections import Counter

def generate_wordclouds():
    n_topics = lda.num_topics

    number_of_words = 300
    # lets first generate the dict of the probabilities for each of the plots
    #
    ndocs_word = news_tfidf.applymap(lambda x: 1 if x > 0 else x).sum()
    #
    coun_vect = CountVectorizer(max_df=0.99, min_df=0.01)
    count_matrix = coun_vect.fit_transform(news_processed_final)
    count_array = count_matrix.toarray()
    counter = pd.DataFrame(data=count_array,columns = coun_vect.get_feature_names_out())
    freq_words = counter.sum()
    #
    corpus = news_processed_final
    frecuencia_palabras = Counter()
    num_documentos = len(corpus)

    for documento in corpus:
        palabras = set(documento.split())
        frecuencia_palabras.update(palabras)
    idf = {}
    for palabra, frecuencia in frecuencia_palabras.items():
        idf[palabra] = math.log((num_documentos + 1) / (frecuencia + 1)) + 1
        
    #start plotting
    titulos = ["Pesos LDA", 
              "Pesos LDA / frecuencia palabras en el corpus", 
              "Pesos LDA / ", 
              "Pesos LDA / idf"]
    for t in range(n_topics):
        # probabilities
        pr = []
        pr.append(dict(lda.show_topic(t, number_of_words)))
        
        probs =lda.show_topic(t,200)
        probs_final1 = dict(lda.show_topic(t,200))
        probs_final2 = dict(lda.show_topic(t,200))
        probs_final3 = dict(lda.show_topic(t,200))
        for i in range(len(probs)): 
            palabra = probs[i][0]
            ndocs = ndocs_word[palabra]    
            probs_final1[palabra] = probs_final1[palabra]/ndocs
            #
            nwords = freq_words[palabra]
            probs_final2[palabra] = probs_final2[palabra]/nwords
            #
            idf_term = idf[palabra]
            probs_final3[palabra] = probs_final3[palabra]/idf_term
            
        pr.append(probs_final1)
        pr.append(probs_final2)
        pr.append(probs_final3)
        
        # Crear figura y ejes
        fig, axes = plt.subplots(2, 2, figsize=(10, 8))
        # Configurar título global
        fig.suptitle("Nube de palabras tópico: #" + str(t) )
        # Generar las cuatro WordClouds
        num=0
        for ax in axes.flatten():
            ax.imshow(WordCloud(background_color="white", random_state=42).fit_words(pr[num]))
            ax.axis("off")
            ax.set_title(titulos[num])
            num=num+1
        # Ajustar el espaciado entre subfiguras
        plt.tight_layout()
        # Guardar la figura creada
        plt.savefig("wordcloud_figures/v1/wordcloud_topic"+str(t)+".png")
        # Mostrar la figura
        plt.show()
        
```

```python
######## TSNE


from sklearn.manifold import TSNE

doc_topic_dist = [lda.get_document_topics(doc, minimum_probability=0) for doc in corpus_news]
doc_topic_matrix = gensim.matutils.corpus2dense(doc_topic_dist, num_terms=lda.num_topics).transpose()
# nombres a los topicos
colormap = {
    0: "#1f77b4",
    1: "#aec7e8",
    2: "#ff7f0e",
    3: "#ffbb78",
    4: "#2ca02c",
    5: "#98df8a",
    6: "#d62728",
    7: "#ff9896",
    8: "#9467bd",
    9: "#c5b0d5",
    10: "#8c564b",
    11: "#c49c94",
    12: "#e377c2",
    13: "#f7b6d2",
    14: "#7f7f7f",
    15: "#c7c7c7",
    16: "#bcbd22",
    17: "#dbdb8d",
    18: "#17becf",
    19: "#9edae5"
}


dominant_topic = [max(topics, key=lambda x: x[1])[0] for topics in doc_topic_dist]
def tsne_plot(iteraciones): 
    p = round(np.sqrt(doc_topic_matrix.shape[0]))
    tsne_model = TSNE(n_components=2, random_state=42,verbose=0, perplexity=p, n_iter=iteraciones)
    tsne_results = tsne_model.fit_transform(doc_topic_matrix)
    
    
    # Plot the results
    fig = plt.figure(figsize=(10, 10))
    markers = [plt.scatter([], [], s=50, color=color, edgecolor='black', label=f'Topic {topic}') for topic, color in colormap.items()]

    plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=[colormap[topic] for topic in dominant_topic])
    plt.legend(handles=markers,bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
    name = "fig_"+str(iteraciones)+".jpg"
    fig.savefig("tsne_figures/v1"+name, dpi=600)


iteraciones = [250,500,1000,5000,10000,20000,50000]
for i in iteraciones: 
    print("Numero iteraciones: ", i)
    tsne_plot(i)
```

```python
# matriz de correlaciones entre las distribuciones de los topicos

from scipy.stats import spearmanr



# Crear una matriz vacía para almacenar las correlaciones
num_topics = lda.num_topics
correlation_matrix = np.zeros((num_topics, num_topics))

# Iterar sobre todos los pares de tópicos y calcular su correlación
for i in range(num_topics):
    for j in range(num_topics):
        if i == j:
            continue
        topic_i = lda.get_topic_terms(i, None)
        topic_j = lda.get_topic_terms(j, None)
        topic_i_words = [id2word[word_id] for word_id, _ in topic_i]
        topic_j_words = [id2word[word_id] for word_id, _ in topic_j]
        common_words = set(topic_i_words).intersection(set(topic_j_words))
        topic_i_weights = [weight for _, weight in topic_i if id2word[_] in common_words]
        topic_j_weights = [weight for _, weight in topic_j if id2word[_] in common_words]
        correlation, _ = spearmanr(topic_i_weights, topic_j_weights)
        correlation_matrix[i][j] = round(correlation,3)



words = [id2word[id] for id in id2word.keys()]
df = pd.DataFrame()

for i in range(lda.num_topics): 
    topic_i = lda.get_topic_terms(i, None)
    topic_i_weights = [weight for _, weight in topic_i if id2word[_] in words]
    
    topic_name = "Tópico "+str(i+1)
    df[topic_name] = topic_i_weights
    
import seaborn as sns

fig, ax = plt.subplots(figsize=(12,12))

sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='Blues', 
           linewidths=1)
```
