```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
tfidf_df = pd.read_csv(r"D:\TFG\tfg_git\data\tf_idf\news_processed_tfidf_fulldataset.csv")
tfidf_df_v3 = pd.read_csv(r"D:\TFG\tfg_git\data\tf_idf\news_processed_tfidfv2_fulldataset.csv")
```

```python
tfidf_df
```

```python
import numpy as np

feature_names = tfidf_df.columns.tolist()

# loop through each row of the matrix and create a list of words for that row
word_lists = []
for row in range(tfidf_df.shape[0]):
    # get the indices of the non-zero elements in the row
    col_indices = np.nonzero(tfidf_df.iloc[row].values)[0]
    # create a list of words for the row
    words = [feature_names[col] for col in col_indices]
    word_lists.append(words)

word_lists_news = word_lists
```

```python
len(word_lists_news)
```

```python
full_list = [item for sublist in word_lists_news for item in sublist]
```

```python
full_list_ngrams = [item for item in full_list if '_' in str(item)]
```

```python
ngrams_count = Counter(full_list_ngrams)
```

```python
sorted_ngrams_count = sorted(ngrams_count.items(), key=lambda x: x[1], reverse=True)
```

```python
for word, count in sorted_ngrams_count[:20]:
    print(f"{word}: {count}")
```

### The top 15 ngrams created with the second case (not proper nouns)

```python
import numpy as np

feature_names = tfidf_df_v3.columns.tolist()

# loop through each row of the matrix and create a list of words for that row
word_lists = []
for row in range(tfidf_df_v3.shape[0]):
    # get the indices of the non-zero elements in the row
    col_indices = np.nonzero(tfidf_df_v3.iloc[row].values)[0]
    # create a list of words for the row
    words = [feature_names[col] for col in col_indices]
    word_lists.append(words)

word_lists_news = word_lists
```

```python
full_list = [item for sublist in word_lists_news for item in sublist]
full_list_ngrams = [item for item in full_list if '_' in str(item)]
ngrams_count = Counter(full_list_ngrams)
sorted_ngrams_count = sorted(ngrams_count.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_ngrams_count[:20]:
    print(f"{word}: {count}")
```
