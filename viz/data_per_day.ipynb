```python
candidates = ["Biden", "Sanders", "Warren", "Bloomberg", "Buttigieg"]
```

```python
news["published_short"].tolist()
```

```python
news.groupby(["published"])["index"].count().index.tolist()
```

```python
news.groupby(["published_short"])["url"].count()
```

```python
series_news = []
series_mails = []

for candidate in candidates: 
    aux = news[news[candidate]!=0]
    aux2 = mails[mails[candidate]!=0]

    t = aux.groupby(["published_short"])["url"].count().index.tolist()
    s1 = aux.groupby(["published_short"])["url"].count()
    s2 = aux2.groupby(["date"])["index"].count()
    t_mais = aux2.groupby(["date"])["index"].count().index.tolist()

    series_news.append(s1)
    series_mails.append(s2)
    
    
fila_nueva = pd.Series({"2020-02-10": 0})
serie_bloomberg = series_mails[3].append(fila_nueva)
serie_bloomberg= serie_bloomberg.sort_index()   
    
```

```python
# create figure and axis objects
fig = plt.figure(figsize=(16, 10))

# add the first subplot
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(t, series_news[0], color='lightblue', label='Biden') 
ax1.plot(t, series_news[1], color='aqua', label='Sanders')
ax1.plot(t, series_news[2], color='green', label='Warren')
ax1.plot(t, series_news[3], color='orange', label='Bloomberg')
ax1.plot(t, series_news[4], color='purple', label='Buttigieg')
ax1.set_xticks(t[::2])
ax1.set_xticklabels(t[::2], fontsize=10)
ax1.set_ylabel('Valor', fontsize=14)
ax1.tick_params('y', labelcolor='black')
ax1.set_title('Notícias', fontsize=20)
# set x-axis tick labels
ax1.tick_params(axis='x', rotation=90, pad=0)
ax1.set_xticks(t[::2])
ax1.set_xticklabels(t[::2], fontsize=14)

# set y-axis labels and colors
ax1.set_ylabel('Número', fontsize=14)
ax1.tick_params('y', labelcolor='black', labelsize=14)

# add the second subplot
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(t_mais, series_mails[0], color='lightblue', label='Biden') 
ax2.plot(t_mais, series_mails[1], color='aqua', label='Sanders')
ax2.plot(t_mais, series_mails[2], color='green', label='Warren')
ax2.plot(t_mais, serie_bloomberg,  color='orange', label='Bloomberg')
ax2.plot(t_mais, series_mails[4], color='purple', label='Buttigieg')
ax2.set_xticks(t[::2])
ax2.set_xticklabels(t[::2], fontsize=10)
ax2.set_ylabel('Valor', fontsize=14)
ax2.tick_params('y', labelcolor='black', labelsize=14)
ax2.set_title('Correos', fontsize=20)
# set x-axis tick labels
ax2.tick_params(axis='x', rotation=90, pad=0)
ax2.set_xticks(t[::2])
ax2.set_xticklabels(t[::2], fontsize=14)

# set y-axis labels and colors
ax2.set_ylabel('Número', fontsize=14)
ax2.tick_params('y', labelcolor='black')

# create a single legend for both subplots
lines, labels = ax1.get_legend_handles_labels()
lgd= ax1.legend(lines, labels, loc='center left', bbox_to_anchor=(0.4, -0.2), ncol=5, fontsize=16)

# adjust spacing between subplots
fig.subplots_adjust(wspace=0.3)

# display the plot
plt.show()

```
