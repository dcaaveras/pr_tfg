```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load data

news = pd.read_csv(r"D:\TFG\tfg_git\data\news\news3_v2.csv")
mails = pd.read_csv(r"D:\TFG\tfg_git\data\mails\corpus_v2.0.csv")


```

```python
s1 = news.groupby(["published_short"])["url"].count()
s2 = mails.groupby(["date"])["index"].count()
t = mails.groupby(["date"])["index"].count().index.tolist()
```

```python


# create figure and axis objects
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# plot the first series on the left axis
ax1.plot(t, s1, 'b-')
ax1.tick_params(axis='x', rotation=90, pad=0)
ax1.set_xticks(t[::2])
ax1.set_xlabel("Dias")
ax1.set_ylabel('Notcias por dia', color='b')
ax1.tick_params('y', colors='b')

# plot the second series on the right axis
ax2.plot(t, s2, 'r-')
ax2.set_ylabel('Mails por da', color='r')
ax2.tick_params('y', colors='r')

# add a title to the figure
plt.title('Notcias/mails por dia')

# display the plot
plt.show()
```
