```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load data

news = pd.read_csv(r"D:\TFG\tfg_git\data\news\news3_v2.csv")
mails = pd.read_csv(r"D:\TFG\tfg_git\data\mails\corpus_v2.0.csv")


```

```python
s1 = news.groupby(["published_short"])["url"].count()
s2 = mails.groupby(["date"])["index"].count()
t = mails.groupby(["date"])["index"].count().index.tolist()
```

```python

# create figure and axis objects
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# plot the first series on the left axis
c1 = "lightblue"
ax1.plot(t, s1, 'b-')
ax1.tick_params(axis='x', rotation=90, pad=0)
ax1.set_xticks(t[::2])
ax1.set_xlabel("Dias")
ax1.set_ylabel('Notícias por dia', color='b')
ax1.tick_params('y', colors='b')

c1 = "aqua"
# plot the second series on the right axis
ax2.plot(t, s2, 'r-')
ax2.set_ylabel('Mails por día', color='r')
ax2.tick_params('y', colors='r')

# add a title to the figure
plt.title('Notícias/mails por dia')

# display the plot
plt.show()
```

```python
import numpy as np
from scipy.signal import savgol_filter

# create figure and axis objects
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# plot the first series on the left axis
c1 = "lightblue"
ax1.plot(t, s1, 'b-', alpha=0.5)  # añade alpha para hacer la linea más clara
ax1.tick_params(axis='x', rotation=90, pad=0)
ax1.set_xticks(t[::2])
ax1.set_xlabel("Dias")
ax1.set_ylabel('Notícias por dia', color='b')
ax1.tick_params('y', colors='b')

# suavizado de la primera serie
s1_smooth = savgol_filter(s1, 5, 2) # parametros: ventana de 5 puntos y grado 2 de polinomio
ax1.plot(t, s1_smooth, 'b-', linewidth=2, alpha=0.5)  # plotea la serie suavizada

# plot the second series on the right axis
c1 = "aqua"
ax2.plot(t, s2, 'r-')  # añade alpha para hacer la linea más clara
ax2.set_ylabel('Mails por día', color='r')
ax2.tick_params('y', colors='r')

# suavizado de la segunda serie
s2_smooth = savgol_filter(s2, 5, 2) # parametros: ventana de 5 puntos y grado 2 de polinomio
ax2.plot(t, s2_smooth, 'r-', linewidth=2, alpha=0.5)  # plotea la serie suavizada

# add a title to the figure
plt.title('Notícias/mails por dia')

# display the plot
plt.show()

```
