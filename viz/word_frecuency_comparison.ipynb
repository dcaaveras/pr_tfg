```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load original data
news = pd.read_csv(r"D:\TFG\tfg_git\data\news\news3_v2.csv")
mails = pd.read_csv(r"D:\TFG\tfg_git\data\mails\corpus_v2.0.csv")
```

```python
# load processed data
df_processed = pd.read_csv(r"D:\TFG\tfg_git\data\clean_text\data_clean_text.csv")
news_processed = df_processed[df_processed["tipo"]=="News"]
mails_processed = df_processed[df_processed["tipo"]=="Mails"].reset_index()
```

```python
len_news_orig = []
len_news_processed = []
for i in range(len(news)): 
    len_news_orig.append(len(news["clean_text"][i]))
    len_news_processed.append(len(news_processed["text_processed"][i]))
```

```python
print(len(len_news_orig))
print(len(len_news_processed))
```

```python
len_mails_orig = []
len_mails_processed = []
for i in range(len(mails)): 
    len_mails_orig.append(len(mails["clean_text"][i]))
    len_mails_processed.append(len(mails_processed["text_processed"][i]))
```

```python
print(len(len_mails_orig))
print(len(len_mails_processed))
```

```python
print(len_news_orig[0:20])
print(len_news_processed[0:20])
```

```python
# crear figura y subplot
fig, ax = plt.subplots(figsize=(8, 5))

# trazar el primer histograma
ax.hist(len_news_orig, bins=10, alpha=0.5, label='Notícias original', color='b', histtype='stepfilled')

# trazar el segundo histograma
ax.hist(len_news_processed, bins=10, alpha=0.5, label='Notícias procesadas', color='r', histtype='stepfilled')

# ajustar las etiquetas y títulos
ax.set_xlabel('Longitud de la cadena')
ax.set_ylabel('Frecuencia')
ax.set_title('Número de carácteres dataset de notícias')
# establecer límites y ajustar las marcas de graduación en el eje x
# establecer límites y ajustar las marcas de graduación en el eje x
ax.set_xlim(0, 60000)
# agregar leyenda
ax.legend()

# mostrar el gráfico
plt.show()
```

```python
# crear figura y subplot
fig, ax = plt.subplots(figsize=(8, 5))

# trazar el primer histograma
ax.hist(len_mails_orig, bins=10, alpha=0.5, label='Notícias original', color='b', histtype='stepfilled')

# trazar el segundo histograma
ax.hist(len_mails_processed, bins=10, alpha=0.5, label='Notícias procesadas', color='r', histtype='stepfilled')

# ajustar las etiquetas y títulos
ax.set_xlabel('Longitud de la cadena')
ax.set_ylabel('Frecuencia')
ax.set_title('Número de carácteres dataset de notícias')
# establecer límites y ajustar las marcas de graduación en el eje x
# establecer límites y ajustar las marcas de graduación en el eje x
ax.set_xlim(0, 10000)
# agregar leyenda
ax.legend()

# mostrar el gráfico
plt.show()
```
