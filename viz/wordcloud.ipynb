```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import find

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from wordcloud import WordCloud
```

```python
candidates = ["Biden", "Sanders", "Warren", "Bloomberg", "Buttigieg"]
```

```python
csv1 = pd.read_csv(r"D:\TFG\tfg_git\data\df_visualizacion\noticias_procesadas_fulldataset.csv")
tfidf1 = pd.read_csv(r"D:\TFG\tfg_git\data\tf_idf\news_processed_tfidf_fulldataset.csv")
```

```python
tfidf1 = tfidf1.drop(["level_0"], axis=1)
```

```python

feature_names = tfidf1.columns.tolist()

# loop through each row of the matrix and create a list of words for that row
word_lists = []
for row in range(tfidf1.shape[0]):
    # get the indices of the non-zero elements in the row
    _, col_indices, _ = find(tfidf1.iloc[row])
    # create a list of words for the row
    words = [feature_names[col] for col in col_indices]
    word_lists.append(" ".join(words))

word_lists_csv1 = word_lists
csv1["words"] = word_lists_csv1
```

```python
csv1[csv1["Biden"]!=0]["words"]
```

```python
# Crear los cinco wordclouds
wordcloud1 = WordCloud(background_color="white", width=1000, height=500, max_words=100).generate(" ".join(csv1[csv1["Biden"]!=0]["words"]))
wordcloud2 = WordCloud(background_color="white", width=1000, height=500, max_words=100).generate(" ".join(csv1[csv1["Sanders"]!=0]["words"]))
wordcloud3 = WordCloud(background_color="white", width=1000, height=500, max_words=100).generate(" ".join(csv1[csv1["Warren"]!=0]["words"]))
wordcloud4 = WordCloud(background_color="white", width=1000, height=500, max_words=100).generate(" ".join(csv1[csv1["Bloomberg"]!=0]["words"]))
wordcloud5 = WordCloud(background_color="white", width=1000, height=500, max_words=100).generate(" ".join(csv1[csv1["Buttigieg"]!=0]["words"]))
```

```python
# Crear la figura y los subplots
fig, axes = plt.subplots(3, 2, figsize=(16, 12))

# Colocar cada wordcloud en un subplot
axes[0, 0].imshow(wordcloud1)
axes[0, 0].set_title('Biden', fontsize=20)
axes[0, 0].axis('off')

axes[0, 1].imshow(wordcloud2)
axes[0, 1].set_title('Sanders', fontsize=20)
axes[0, 1].axis('off')

axes[1, 0].imshow(wordcloud3)
axes[1, 0].set_title('Warren', fontsize=20)
axes[1, 0].axis('off')

axes[1, 1].imshow(wordcloud4)
axes[1, 1].set_title('Bloomberg', fontsize=20)
axes[1, 1].axis('off')

axes[2, 0].imshow(wordcloud5)
axes[2, 0].set_title('Buttigieg', fontsize=20)
axes[2, 0].axis('off')

# Eliminar el subplot vaco
axes[2, 1].axis('off')

plt.tight_layout()
plt.subplots_adjust(hspace=0.2)


# Ajustar la posicin de los subplots de la segunda fila
pos1 = axes[1, 0].get_position()
pos2 = axes[1, 1].get_position()
pos_orig = axes[2, 0].get_position()
pos_merged = [(pos1.x0 + pos1.x1)/2, 
              pos_orig.y0, 
              (pos2.x0 + pos2.x1)/2, 
              pos_orig.y1]

pos_merged = [0.1, 
              pos_orig.y0, 
              (pos2.x0 + pos2.x1)/2, 
              pos_orig.y1]
axes[2, 0].set_position(pos_merged)
# Mostrar la figura
fig.set_figheight(15)
fig.set_figwidth(15)
plt.show()
```
